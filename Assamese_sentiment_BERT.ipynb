```python
import pandas as pd
from deep_translator import GoogleTranslator


df = pd.read_csv('Assamese Sentiments file.csv')


first_column_values = df.iloc[0:100, 0].tolist()
sentences_translated = []  # List to store translated sentences

# Translate each sentence and append to the list
for sentence in first_column_values:
    translated_text = GoogleTranslator(source='auto', target='en').translate(sentence)
    sentences_translated.append(translated_text)


for translated_sentence in sentences_translated:
    print(translated_sentence)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import requests
from bs4 import BeautifulSoup
import re
```

```python
import sentencepiece
```

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
tokens=[]
for lines in sentences_translated:
    tokens.append(tokenizer.encode(lines, return_tensors='pt'))
```

```python
results=[]
for i in tokens:
    results.append(model(i))
```

```python
rating= []
for i in results:
    rating.append(int(torch.argmax(i.logits))+1)
```

```python
#Note that a sentiment rating of 1 means "Very negative" whereas a rating of 5 means "Very Positive". All other sentiments lie between 1 and 5
```

```python
for i in range (0,20,1):
    print("Sentence"+str(i))
    print("\n")
    print(first_column_values[i])
    print("\n")
    print("English Translation : ")
    print(sentences_translated[i])
    print("\n")
    print("Rating : ")
    print(rating[i])
    print("\n")
    print("\n")
```
